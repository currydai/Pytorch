### 功率谱密度

#### 1.调用matpoltlib中psd函数

```python
#matplotlib auto
fig1=plt.figure()
plt.psd(data.iloc[:,0], NFFT=10000,window=mlab.window_hanning,Fs=300)
plt.xlim(0,300)
#plt.ylim(-0.1,0.1)
plt.xscale('symlog')
plt.yscale('symlog')
plt.show()
```

#### 2.调用sklearn中signal.welch函数

使用韦尔奇方法估算功率谱密度
用法说明：https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.welch.html

```python
#matplotlib auto
fig1=plt.figure()
f, Pxx_den = signal.welch(data.iloc[:,0]*17.6,300, nperseg=10000)
plt.plot(f,np.sqrt(Pxx_den))
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Hz')
plt.ylabel('[nT/sprt(Hz)]')
plt.grid(linestyle='-.')
plt.legend(['Bx'])
plt.show()
```

结论：方法1与方法2所成的功率谱密度一致,将方法2封装为功率谱密度函数

```python
def plt_welch(sig):
    f, Pxx_den = signal.welch(sig, nperseg=10000)
    return Pxx_den
```

### 互谱

使用韦尔奇方法估计交叉功率谱密度Pxy
用法说明：https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.csd.html

```python
f, Pxy = signal.csd(data.iloc[:,0]*17.6, data.iloc[:,3]*17.6, 300, nperseg=10000)
temp=np.abs(Pxy)
fig2=plt.figure()
plt.semilogy(f, np.sqrt(temp))
plt.xlabel('frequency [Hz]')
plt.ylabel('CSD [nT/Hz]')
plt.grid(linestyle='-.')
plt.legend(['Bx-bx'])
plt.xscale('log')
plt.show()
```

将上述方法封装为函数，用来后续调用：

```python
def plt_csd(sig1,sig2):
    f, Pxy = signal.csd(sig1, sig2, 300, nperseg=10000)
    return [f,np.abs(Pxy)]
```

### 互谱自谱画在一起

```python
def plot_all(sig1,sig2):
    f, Pxx_den = signal.welch(sig1,300, nperseg=10000)
    f, Pxy = signal.csd(sig1,sig2, 300, nperseg=10000)
    temp=np.abs(Pxy)
    
    fig3=plt.figure()
    
    #功率谱
    plt.subplot(2,1,1)
    plt.plot(f,np.sqrt(Pxx_den))
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel('Hz')
    plt.ylabel('[nT/sprt(Hz)]')
    plt.grid(linestyle='-.')
    plt.legend(['Bx'])
    
    #自谱
    plt.subplot(2,1,2)
    plt.semilogy(f, np.sqrt(temp))
    plt.xlabel('frequency [Hz]')
    plt.ylabel('CSD [nT/Hz]')
    plt.grid(linestyle='-.')
    plt.legend(['Bx-bx'])
    plt.xscale('log')
    plt.show()
    
    fig4=plt.figure()
    plt.plot(f,np.sqrt(Pxx_den))
    plt.semilogy(f, np.sqrt(temp))
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel('Hz')
    plt.ylabel('[nT/sprt(Hz)]')
    plt.grid(linestyle='-.')
    plt.legend(['Bx','Bx-bx'])
    plt.show()
```

```python
plot_all(data.iloc[:,0]*17.6, data.iloc[:,3]*17.6)
```

结论：相同方向的轴自谱和互谱一致(Bx-Bx)

```python
plot_all(data.iloc[:,0]*17.6, data.iloc[:,1]*17.6)
```

结论：不同方向的轴自谱和互谱有差别(Bx-By)

### 计算$S_i=\sqrt{S_{aa}S_{bb}}- |{S_{ab}}|$

计算本征噪声功率的估计值

```python
def Si(sig1,sig2):
    f, Pxx_den_a = signal.welch(sig1,300, nperseg=10000)
    f, Pxx_den_b = signal.welch(sig1,300, nperseg=10000)
    f, Pxy_ab = signal.csd(sig1, sig2, 300, nperseg=10000)
    
    Saa=Pxx_den_a
    Sbb=Pxx_den_b
    
    Sab=np.abs(Pxy_ab)
    
    S_i=np.sqrt(np.sqrt(Saa)*np.sqrt(Sbb))-np.sqrt(Sab)
    
    fig4=plt.figure()
    plt.plot(f,S_i)
    plt.xlabel('frequency [Hz]')
    plt.ylabel('CSD [nT/Hz]')
    plt.grid(linestyle='-.')
    plt.legend(['Bx-bx'])
    plt.xscale('log')
    plt.yscale('log')
    plt.show()
    
    return S_i
```

```python
S_i=Si(data.iloc[:,0]*17.6, data.iloc[:,3]*17.6)
```

```python
f, Pxx_den = signal.welch(data.iloc[:,0]*17.6,300, nperseg=10000)
f, Pxx_den_buchanghou = signal.welch(data_2.iloc[:,0]*17.6,300, nperseg=10000)
```

```python
fig5=plt.figure()
plt.plot(f,np.sqrt(Pxx_den))
plt.plot(f,S_i)
plt.plot(f,np.sqrt(Pxx_den_buchanghou))
plt.xlabel('frequency [Hz]')
plt.ylabel('CSD [nT/Hz]')
plt.grid(linestyle='-.')
plt.legend(['Bx','Bx-bx','Bx_buchanghou'])
plt.xscale('log')
plt.yscale('log')
plt.show()
```

* **Bx:原始信号的功率谱密度**
* **Bx_bx:本征噪声功率谱密度**
* **Bx_buchanghou：补偿后的功率谱密度**
